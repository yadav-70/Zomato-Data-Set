```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# To comibe two different datafromes we have to use merge function
#We are merging the data fromae on country code and the merge values will appear on left right side of the 
pd.merge(df,df_country,on='Country Code',how='left')
```

```python
#Here we are assign the megred value to another df
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
#Checking the ewhich country has maximum order placed through Zomato 
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#Pie Charts  top 5 countries 
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observations: Zomato maximum records  or transaction are from India,USA and then UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
ratings
```

#Observation 
1. When rating is between 4.5 to 4.9 is Excellent 
2. When ratings are beteen 4.0 to 4.4is very good
3. When ratings are beteen 3.5 to 3.9 is  good
4. When ratings are beteen 2.5 to 3.4is average
5. When ratings are beteen 1.8 to 2.4 is  poor


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color', data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```

# Observation 
1. Not rated count is very high
2. Maximum numbers of rating are between 2.5 to 3.4 

```python
#Count Plot 
sns.countplot(x='Rating color',data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```
